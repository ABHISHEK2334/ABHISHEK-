```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
import numpy as np

# Create some example data with missing values
X = np.array([[1, 2, np.nan],
              [3, np.nan, 4],
              [np.nan, 5, 6]])   #chatgpt

# Initialize IterativeImputer (MICE) with additional parameters
imputer = IterativeImputer(max_iter=10, random_state=0)

# Fit the imputer model on the data
imputer.fit(X)

# Transform the data to fill in missing values
X_imputed = imputer.transform(X)

print("Imputed Data:\n", X_imputed)
```
