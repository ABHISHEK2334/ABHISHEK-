```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

# Setting a random seed for reproducibility
np.random.seed(23)

# Generating two classes of samples
mu_vec1 = np.array([0, 0, 0])
cov_mat1 = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
class1_sample = np.random.multivariate_normal(mu_vec1, cov_mat1, 20)

mu_vec2 = np.array([1, 1, 1])
cov_mat2 = np.array([[1, 0, 0], [0, 1, 0], [0, 0, 1]])
class2_sample = np.random.multivariate_normal(mu_vec2, cov_mat2, 20)

# Creating DataFrames and concatenating
df = pd.DataFrame(class1_sample, columns=['feature1', 'feature2', 'feature3'])
df['target'] = 1
df1 = pd.DataFrame(class2_sample, columns=['feature1', 'feature2', 'feature3'])
df1['target'] = 0
df = pd.concat([df, df1], ignore_index=True)
df = df.sample(40).reset_index(drop=True)

# Plotting the original data in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(df['feature1'], df['feature2'], df['feature3'], c=df['target'], cmap='coolwarm')
legend1 = ax.legend(*scatter.legend_elements(), title="Classes")
ax.add_artist(legend1)
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Feature 3')

# Standardizing the features
scaler = StandardScaler()
df.iloc[:, 0:3] = scaler.fit_transform(df.iloc[:, 0:3])

# Computing the covariance matrix
cov_matrix = np.cov(df.iloc[:, 0:3].T)

# Computing eigenvalues and eigenvectors
eigen_values, eigen_vectors = np.linalg.eig(cov_matrix)

# Sorting eigenvalues and eigenvectors
sorted_index = np.argsort(eigen_values)[::-1]
sorted_eigenvalues = eigen_values[sorted_index]
sorted_eigenvectors = eigen_vectors[:, sorted_index]

# Selecting the top 2 eigenvectors (principal components)
eigenvector_subset = sorted_eigenvectors[:, 0:2]

# Transforming the data
X_reduced = np.dot(df.iloc[:, 0:3], eigenvector_subset)

# Creating a DataFrame for the reduced data
df_pca = pd.DataFrame(X_reduced, columns=['PC1', 'PC2'])
df_pca['target'] = df['target']

# Plotting the reduced data
plt.figure(figsize=(8, 6))
scatter = plt.scatter(df_pca['PC1'], df_pca['PC2'], c=df_pca['target'], cmap='coolwarm')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of the Dataset')
legend1 = plt.legend(*scatter.legend_elements(), title="Classes")
plt.gca().add_artist(legend1)
plt.show()

# Drawing 3D arrows for principal components
ax.quiver(0, 0, 0, sorted_eigenvectors[0, 0], sorted_eigenvectors[1, 0], sorted_eigenvectors[2, 0],
          color='r', label='PC1', length=3, normalize=True)
ax.quiver(0, 0, 0, sorted_eigenvectors[0, 1], sorted_eigenvectors[1, 1], sorted_eigenvectors[2, 1],
          color='g', label='PC2', length=3, normalize=True)
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


np.random.seed(42)
X = np.random.rand(100, 1) - 0.5
y = 3*X[:, 0]**2 + 0.05 * np.random.randn(100)


import pandas as pd


df = pd.DataFrame()


df['X'] = X.reshape(100)
df['y'] = y


df
```

```python
plt.scatter(df['X'], df['y'])
plt.show()
```

```python
df['pred1']=df['y'].mean()
```

```python
df['res1']=df['y']-df['pred1']
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree1=DecisionTreeRegressor(max_leaf_nodes=8)
tree1.fit(df['X'].values.reshape(100,1),df['res1'].values)
```

```python
from sklearn.tree import plot_tree
plot_tree(tree1)
plt.show()
```

```python
X_test = np.linspace(-0.5, 0.5, 500)
y_pred = 0.265458 + tree1.predict(X_test.reshape(500, 1))
```

```python
plt.figure(figsize=(14,4))
plt.subplot(121)
plt.plot(X_test, y_pred, linewidth=2,color='red')
plt.scatter(df['X'],df['y'])
```
