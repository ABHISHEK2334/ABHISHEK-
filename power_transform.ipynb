````python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PowerTransformer
````

````python
path = "/content/concrete-data.csv"
df=pd.read_csv(path,encoding='latin-1')
df.head()
````

````python
x=df.drop(columns=['Strength'])
y=df.iloc[:,-1]
````

````python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
````

````python
ir=LinearRegression()
ir.fit(x_train,y_train)# Applying Regression without any transformation
y_pred=ir.predict(x_test)
r2_score(y_test,y_pred)
````

````python
ir=LinearRegression()
np.mean(cross_val_score(ir,x,y,scoring='r2'))
````

````python
# Plotting the distplots without any transformation
for col in x_train.columns:
  plt.figure(figsize=(14,4))
  plt.subplot(121)
  sns.distplot(x_train[col])

  plt.subplot(122)
  stats.probplot(x_train[col],dist="norm",plot=plt)
  plt.show()
````

````python
pt=PowerTransformer(method='box-cox')
x_tain_transformed =pt.fit_transform(x_train+0.000001)
x_test_transformed=pt.fit_transform(x_test+0.000001)
pd.DataFrame({'cols':x_train.columns,'lambdas':pt.lambdas_})
````

````python
ir=LinearRegression()
ir.fit(x_train_transformed,y_train)
y_pred2=ir.predict(x_train_transformed)
r2_score=(y_test,y_pred2)
````

````python
pt=PowerTransformer(method="box-cox")
x_transformed=pt.fit_transform(x+0.00000001)
ir=LinearRegression()
np.mean(cross_val_score(ir,x_transformed,y,scoring='r2') )
````

````python
x_train_transformed=pd.DataFrame(x_train_transformed,columns=x_train.columns)
for col in x_train_transformed.columns:

  plt.figure(figsize=(14,4))
  plt.subplot(121)
  sns.distplot(x_train[col])


  plt.subplot(122)
  sns.distplot(x_train_transformed[col])


````

````python
pt1=PowerTransformer()
x_train_transformed2=pt1.fit_transform(x_train)
x_test_transformed2=pt1.fit_transform(x_test)

ir=LinearRegression()
ir.fit (x_train_transformed2,y_train )

y_pred3=ir.predict(x_test_transformed2)
pd.DataFrame({"colns":x_train.columns,"Yeo_Johnson_lambdas":pt1.lambdas_})
````
