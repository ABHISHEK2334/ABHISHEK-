```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('iris.csv')
df.head()
```

```python
df['Species'].values
x = df.drop(columns=['Species'])
y = pd.factorize(df['Species'])[0]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)

```

```python
y
```

```python
clf=RandomForestClassifier(n_jobs=2,random_state=0)
clf1=clf.fit(x_train,y_train)


```

```python
y_pred=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
# Visualizing an individual decision tree within the random forest
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

# Accessing one of the decision trees from the random forest
estimator = clf1.estimators_[0]  # Change the index as needed to access different trees

plt.figure(figsize=(12, 8))
plot_tree(estimator, feature_names=x.columns, class_names=df['Species'].unique(), filled=True)
plt.show()
```
